```python
#1.	customer’s age that makes more purchase
```

```python
Byear = df['Year_Birth']
Age = 2023 - Byear
df['Age'] = Age
```

```python
age_purch = df.groupby(['Age','TotalPurchase']).sum().reset_index()
purchase_age = age_purch[['Age','TotalPurchase']].sort_values('TotalPurchase', ascending = False)
purchase_age.head()
```

```python
#2.	what is the highest purchase made with respect to education
```

```python
education_purch = df.groupby('Education').sum().reset_index()
education_purch[['Education','TotalPurchase']].sort_values('TotalPurchase', ascending = False)
```

```python
#3.	what is the highest purchase made with respect to marital status
```

```python
marital_status_purch = df.groupby('Marital_Status').sum().reset_index()
marital_status_purch[['Marital_Status','TotalPurchase']].sort_values('TotalPurchase', ascending = False)
```
